# Тема проекта: HR Аналитика для прогнозирования статуса сотрудника

## Описание проекта

HR-аналитики компании «Работа с заботой» помогают бизнесу оптимизировать управление персоналом: бизнес предоставляет данные, а аналитики предлагают, как избежать финансовых потерь и оттока сотрудников. В этом HR-аналитикам пригодится машинное обучение, с помощью которого получится быстрее и точнее отвечать на вопросы бизнеса.  
Компания предоставила данные с характеристиками сотрудников компании. Среди них — уровень удовлетворённости сотрудника работой в компании. Эту информацию получили из форм обратной связи: сотрудники заполняют тест-опросник, и по его результатам рассчитывается доля их удовлетворённости от 0 до 1, где 0 — совершенно неудовлетворён, 1 — полностью удовлетворён. 
Собирать данные такими опросниками не так легко: компания большая, и всех сотрудников надо сначала оповестить об опросе, а затем проследить, что все его прошли.

```python
# Функция для вывода информации о дубликатах
def print_duplicates(df_dict):
    for name, df in df_dict.items():
        print(f"Явные дубликаты в датафрейме {name}:")
        duplicates = df[df.duplicated()]
        if not duplicates.empty:
            print(duplicates)
        else:
            print("Явных дубликатов нет")
        print("\n")

# Вызов функции для вывода дубликатов
print_duplicates(dataframes)
```

```python
# Функция для нахождения неявных дубликатов
def find_implicit_duplicates(df_dict):
    for name, df in df_dict.items():
        print(f"Неявные дубликаты в датафрейме {name}:")
        
        #приведение строковых колонок к нижнему регистру и удаление пробелов
        df_cleaned = df.apply(lambda col: col.map(lambda x: x.strip().lower() if isinstance(x, str) else x))
        
        #нахождение дубликатов
        duplicates = df_cleaned[df_cleaned.duplicated()]
        
        if not duplicates.empty:
            print(duplicates)
        else:
            print("Нет неявных дубликатов")
        print("\n")

# Вызов функции для нахождения неявных дубликатов
find_implicit_duplicates(dataframes)
```

```python
# Функция для построения гистограммы распределения
def plot_histogram(df, column):
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], kde=True)
    plt.title(f'Гистограмма распределения для категории {column}')
    plt.xlabel(column)
    plt.ylabel('Частота')
    plt.show()

# Функция для построения гистограммы размаха (boxplot)
def plot_boxplot(df, column):
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[column])
    plt.title(f'Гистограмма размаха для категории {column}')
    plt.xlabel(column)
    plt.show()

# Функция для построения частоты встречаемости для дискретных признаков
def plot_countplot(df, column):
    plt.figure(figsize=(10, 6))
    sns.countplot(x=df[column])
    plt.title(f'Частота встречаемости для категории {column}')
    plt.xlabel(column)
    plt.ylabel('Частота')
    plt.show()

# Функция для построения круговой диаграммы для категориальных признаков
def plot_pie_chart(df, column):
    plt.figure(figsize=(10, 6))
    df[column].value_counts().plot(kind='pie', autopct='%1.1f%%')
    plt.title(f'Круговая диаграмма для категории {column}')
    plt.ylabel('')
    plt.show()
```

```python
# Функция для построения всех графиков для датафрейма
def plot_all_graphs(df):
    print("Построение графиков для датафрейма:")
    for column in df.columns:
        if column == 'id':
            continue
        if pd.api.types.is_numeric_dtype(df[column]):
            unique_values = df[column].nunique()
            if unique_values < 20:  # Условие для дискретных признаков
                print(f"Построение графиков для дискретного признака {column}:")
                plot_countplot(df, column)
            else:  # Условие для непрерывных признаков
                print(f"Построение графиков для непрерывного признака {column}:")
                plot_histogram(df, column)
                plot_boxplot(df, column)
        else:
            print(f"Построение графиков для категориального признака {column}:")
            plot_pie_chart(df, column)
```

```python
# Вызов функции для построения всех графиков для каждого датафрейма
plot_all_graphs(dataframes['train_job_satisfaction_rate'])
```

```python
# Разделение категориальных признаков на те, которые будут кодироваться OneHotEncoder и OrdinalEncoder
onehot_features = ['dept']
ordinal_features = ['level', 'workload', 'last_year_promo', 'last_year_violations']
```

```python
# Пайплайн для категориальных признаков, которые будут кодироваться OneHotEncoder
onehot_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', drop='first'))
])
```

```python
# Пайплайн для категориальных признаков, которые будут кодироваться OrdinalEncoder
ordinal_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ordinal', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])
```

```python
# Пайплайн для числовых признаков
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

# Объединение пайплайнов в ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('onehot', onehot_transformer, onehot_features),
        ('ordinal', ordinal_transformer, ordinal_features)
    ])
```

```python
# Функция SMAPE (англ. symmetric mean absolute percentage error, «симметричное среднее абсолютное процентное отклонение»)
def smape(y_true, y_pred):
    return 100 * np.mean(2 * np.abs(y_pred - y_true) / (np.abs(y_true) + np.abs(y_pred)))
```

```python
# Кастомный скоррер для SMAPE
smape_scorer = make_scorer(smape, greater_is_better=False)
```

```python
# Пайплайн для линейной модели
linear_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', LinearRegression())
])

# Оценка линейной модели на кросс-валидации
linear_scores = cross_val_score(linear_pipeline, X_train, y_train, cv=5, scoring=smape_scorer)
print(f"SMAPE для линейной модели на кросс-валидации: {-linear_scores.mean():.2f}")
```

```python
# Пайплайн для дерева решений
tree_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', DecisionTreeRegressor(random_state=42))
])
```

```python
# Параметры для подбора гиперпараметров дерева решений
param_grid = {
    'model__max_depth': [3, 5, 7, 10],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [1, 2, 4]
}

# GridSearchCV для подбора гиперпараметров дерева решений
grid_search = GridSearchCV(estimator=tree_pipeline, param_grid=param_grid, cv=5, scoring=smape_scorer, n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
# Лучшая модель после подбора гиперпараметров
best_tree_model = grid_search.best_estimator_

# Вывод результатов
print(f"SMAPE для лучшей модели дерева решений на кросс-валидации: {-grid_search.best_score_:.2f}")
print(f"Лучшие гиперпараметры для дерева решений: {grid_search.best_params_}")
```

Нулевая гипотеза (H0): Нет статистически значимого различия в распределениях уровня удовлетворённости уволившихся и оставшихся сотрудников.  

Альтернативная гипотеза (H1): Есть статистически значимое различие в распределениях уровня удовлетворённости уволившихся и оставшихся сотрудников.

```python
# t-тест для сравнения распределений уровня удовлетворённости
t_stat, p_value = ttest_ind(satisfaction_quit, satisfaction_stay, equal_var=False)

print(f"t-статистика: {t_stat:.2f}")
print(f"p-значение: {p_value:.2e}")

if p_value < 0.05:
    print("Различие в распределениях уровня удовлетворённости уволившихся и оставшихся сотрудников статистически значимо.")
else:
    print("Различие в распределениях уровня удовлетворённости уволившихся и оставшихся сотрудников не является статистически значимым.")
```

Распределение уровня удовлетворённости уволившихся сотрудников смещено влево, что указывает на более низкие значения удовлетворённости.  
Распределение уровня удовлетворённости оставшихся сотрудников более равномерно и смещено вправо, что указывает на более высокие значения удовлетворённости.  

Отрицательное значение t-статистики указывает на то, что средний уровень удовлетворённости уволившихся сотрудников значительно ниже, чем у оставшихся сотрудников.

p-значение (2.77e-101) значительно меньше 0.05, что указывает на статистически значимое различие между распределениями уровня удовлетворённости уволившихся и оставшихся сотрудников.

Различие в распределениях уровня удовлетворённости уволившихся и оставшихся сотрудников статистически значимо.
Это подтверждает утверждение аналитиков о том, что уровень удовлетворённости сотрудника работой в компании влияет на вероятность его увольнения.

```python
# Определение числовых признаков, включая новый признак, предсказанный моделью из предыдущей задачи
numerical_features_quit = ['employment_years', 'supervisor_evaluation', 'salary', 'predicted_job_satisfaction_rate']
```

```python
# Разделение категориальных признаков на те, которые будут кодироваться OneHotEncoder и OrdinalEncoder
onehot_features_quit = ['dept', 'last_year_promo', 'last_year_violations']
ordinal_features_quit = ['level', 'workload']
```

```python
# Пайплайн для категориальных признаков, которые будут кодироваться OneHotEncoder
onehot_transformer_quit = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', drop='first'))
])

# Пайплайн для категориальных признаков, которые будут кодироваться OrdinalEncoder
ordinal_transformer_quit = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ordinal', OrdinalEncoder(categories=[
        ['junior', 'middle', 'senior'],
        ['medium', 'high', 'low']
    ], handle_unknown='use_encoded_value', unknown_value=-1))
])

# Пайплайн для числовых признаков
numerical_transformer_quit = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])
```

```python
# Объединение пайплайнов в ColumnTransformer
preprocessor_quit = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer_quit, numerical_features_quit),
        ('onehot', onehot_transformer_quit, onehot_features_quit),
        ('ordinal', ordinal_transformer_quit, ordinal_features_quit)
    ])
```

```python
X_train_quit_prepared = preprocessor_quit.fit_transform(X_train_quit_with_job_satisfaction)
X_test_quit_prepared = preprocessor_quit.transform(X_test_quit)
```

### 6. Обучение модели

```python
# Модель 1: Логистическая регрессия
logreg = LogisticRegression(random_state=42)
roc_auc_logreg = cross_val_score(logreg, X_train_quit_prepared, y_train_quit, cv=5, scoring='roc_auc').mean()

# Модель 2: Случайный лес
rf = RandomForestClassifier(random_state=42)
param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}
grid_search_rf = GridSearchCV(rf, param_grid_rf, cv=5, scoring='roc_auc', n_jobs=-1)
grid_search_rf.fit(X_train_quit_prepared, y_train_quit)
best_rf = grid_search_rf.best_estimator_
roc_auc_rf = grid_search_rf.best_score_

# Модель 3: SVM
svc = SVC(probability=True, random_state=42)
param_grid_svc = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf']
}
grid_search_svc = GridSearchCV(svc, param_grid_svc, cv=5, scoring='roc_auc', n_jobs=-1)
grid_search_svc.fit(X_train_quit_prepared, y_train_quit)
best_svc = grid_search_svc.best_estimator_
roc_auc_svc = grid_search_svc.best_score_

print(f"ROC-AUC для Логистической регрессии: {roc_auc_logreg:.4f}")
print(f"ROC-AUC для Случайного леса: {roc_auc_rf:.4f}")
print(f"ROC-AUC для SVM: {roc_auc_svc:.4f}")
```

```python
# Выбор лучшей модели
best_model = None
best_roc_auc = 0
if roc_auc_logreg > best_roc_auc:
    best_model = logreg
    best_roc_auc = roc_auc_logreg
if roc_auc_rf > best_roc_auc:
    best_model = best_rf
    best_roc_auc = roc_auc_rf
if roc_auc_svc > best_roc_auc:
    best_model = best_svc
    best_roc_auc = roc_auc_svc

print(f"Лучшая модель: {best_model.__class__.__name__} с ROC-AUC: {best_roc_auc:.4f}")
```

```python
# Применение предобработки к тестовым данным
X_test_quit_prepared = preprocessor_quit.transform(X_test_quit)

# Предсказание вероятностей классов с использованием лучшей модели
y_pred_best_model = best_model.predict_proba(X_test_quit_prepared)[:, 1]

# Вычисление ROC-AUC на тестовой выборке
roc_auc_best_model = roc_auc_score(y_test_quit, y_pred_best_model)

# Вывод значения ROC-AUC для лучшей модели на тестовой выборке
print(f"ROC-AUC для лучшей модели на тестовой выборке: {roc_auc_best_model:.4f}")
```
